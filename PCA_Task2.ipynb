```python
# After subspace selection, we decide to use Top 12 components for modeling normal subspace 
# and next 3 components for anomalous subspace

pca = decomposition.PCA(n_components=15)
pca.fit(normalized_training)
pca_model = pca.transform(normalized_training)
```

```python
#training_2 is training dataset 2
training_2  = pd.read_csv('C:\\Users\\Aishwarya\\notebooks\\CDA\\assignment 2\\BATADAL_dataset04.csv', parse_dates=True, index_col='DATETIME')
```

```python
#we set the labels as the col ATT_FLAG to predict and check our accuracy
labels=training_2[' ATT_FLAG']
print(labels)
```

```python
del training_2[' ATT_FLAG']
```

```python

normalized_test  = scaler.fit_transform(training_2)
pca_test = pca.transform(normalized_test)
```

```python
eigenvectors = pca.components_
print(eigenvectors)
# Matrix P represents principal components corresponding to normal subspace
P = np.transpose(eigenvectors[:-3])
P_T = np.transpose(P)
C = np.dot(P, P_T)

# Identity Matrix with dimensions 43 X 43
I = np.identity(43)

# y_residual is the projection of test data on anomalous subspace
y_residual = np.zeros((normalized_test.shape))

# Calculate projection of test data on anomalous subspace
for i in range(normalized_test.shape[0]):
    # Convert row to column vector
    y = np.transpose(normalized_test[i])
    y_residual[i] = np.dot(I - C, y)
```

```python
#calculating spe and threshold has been obtained by calculation as mentioned in pca_threshold
spe = np.zeros((normalized_test.shape[0]))

# na will be set to 1 if the spe is greater than the threshold 
na = np.zeros((normalized_test.shape[0]))
threshold = 147
for i in range(normalized_test.shape[0]):
    spe[i] = np.square(np.sum(np.subtract(y_residual[i], normalized_test[i])))     
    
    # if spe is greater than threshold then classify as attack by setting na to 1
    if(spe[i] > threshold):
        na[i] = 1
    print(na[i]) 

```

```python
#we plot residual vector for our training set 2 which has been used as a test set in our task
training_2=training_2.assign(ResidualVector=spe)
training_2['ResidualVector'].plot(figsize=(15,5))
```

```python
#getting the confusion matrix for our task
tp = 0
fp = 0
tn = 0
fn = 0


for i in range(normalized_test.shape[0]):
    
    if(labels[i] == 1 and na[i] == 1.0):
        tp = tp + 1
    if(labels[i] == -999 and na[i] == 1.0):
        fp = fp + 1
    if(labels[i] == -999 and na[i] == 0.0):
        fn = fn + 1    
    if(labels[i] == 1 and na[i] == 0.0):
        tn = tn + 1
print (tp)
print (fp)
print (fn)
print (tn)

```

```python
#calcualting accuracy, precision and recall
```

```python
print( 'Accuracy:  ', 1.0 * (tp+tn)/(tp+fp+tn+fn))
print ('precision: ' ,1.0 * tp/(tp+fp))
print ('recall:    ' ,1.0 * tp/(tp+fn))
```
